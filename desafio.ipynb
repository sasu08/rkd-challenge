#### Casualmente son de la misma saga, y googleando podemos ver que la duracion de los titulos coincide con lo que esta en la columna rating, por lo que vamos a proceder a mover ese contenido a la columna correspondiente.

#### Como son pocas, podemos hacerlo de forma manual.

```python
df_netflix['duration'][5541] = df_netflix['rating'][5541]
df_netflix['rating'][5541] = np.nan

df_netflix['duration'][5794] = df_netflix['rating'][5541]
df_netflix['rating'][5794] = np.nan

df_netflix['duration'][5813] = df_netflix['rating'][5541]
df_netflix['rating'][5813] = np.nan
```

```python
df_netflix['release_year'].value_counts(ascending=True) 
```

```python
df_netflix = df_netflix.drop(df_netflix.loc[df_netflix.loc[:, 'release_year'] == '40 min'].index)
df_netflix.reset_index(drop=True, inplace=True)
```

```python
df_netflix.release_year = df_netflix.release_year.fillna(0.0).astype(int)
```

### Observaciones

#### - Podemos observar que hay varias filas con algun campo NaN, dependiendo lo que quisieramos analizar deberiamos realizar una limpieza de esos campos o simplemente dejarlos.

#### - En la columna duracion no hay un formato determinado. Se dividen en duracin por minutos y duracin por temporadas segn sea pelcula o serie, correspondientemente. Si quisieramos realizar un anlisis de duracin obligadamente deberamos analizar las series por un lado y las pelculas por otro. Tampoco puede normalizarse porque no podemos pasar temporadas a minutos ni minutos a temporadas.

#### - A parte de type movie y tv show, existe un type William Wyler que muy probablemente se trate de un error.

#### - Parseamos la columna release_year a int y descubrimos que hay un valor "infiltrado" (40 min). Casualmente coincide con la fila que tenia el type = William Wyler.

#### - En principio, no se repite ningun titulo. Puede suceder que si se repitan pero no coincida la totalidad de los caracteres.

#### - Al analizar los valores de la columna 'rating' podemos observar que, por lo menos 3 filas, poseen un error. Donde deberia ir el rating de ese titulo, encontramos la duracion.

#### - Hay peliculas que se repiten su descripcion. Se observa que en la mayoria de los casos son la misma pelicula pero en idiomas/versiones distintas.
